```python
import pandas as pd
```

```python
df = pd.read_csv('sales_dataset.csv')
df
```

```python
#rename the columns properly
```

```python
df = df.rename(columns = {'Customer ID':'Customer_ID', 'Sales Amount':'Sales_Amount', 'Date of Purchase':'Date_of_Purchase'})
df.head()
```

```python
#Check for null values and handle it
```

```python
df.isnull().sum()
```

```python
df[df['Name'].isna() == 1]
```

```python
df.dropna(subset = ["Name", "Customer_ID"], inplace = True)    #drop the entries with no name or customer Id as they are essential values and filling in wrong values can lead to utliers
df[df['Name'].isna() == 1]
```

```python
df['Sales_Amount'] = df['Sales_Amount'].fillna(df['Sales_Amount'].median())    #fill missing values of numerical column with median/mean
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mode()[0])     #fill missing values of categorical column with mode
```

```python
df.isnull().sum()
```

```python
#Remove duplicates
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
#Standardizing the text
```

```python
df["Gender"] = df["Gender"].str.lower().replace({
    "m": "Male", 
    "male": "Male", 
    "M": "Male",
    "f": "Female", 
    "female": "Female", 
    "F":"Female"
})
```

```python
df["Country"] = df["Country"].str.lower()

df["Country"] = df["Country"].replace({
    "usa": "USA",
    "u.k.": "UK",
    "uk": "UK",
    "france": "France",
    "india": "India"
})
```

```python
df
```

```python
#Date Formats
```

```python
df["Date_of_Purchase"] = pd.to_datetime(df["Date_of_Purchase"], dayfirst=True)

df["Date_of_Purchase"] = df["Date_of_Purchase"].dt.strftime("%d-%m-%Y")     #to convert to dd-mm-yyyy formt
```

```python
df
```

```python
#checking dtypes
```

```python
df.dtypes
```

```python
df["Age"] = df["Age"].replace("Thirty-five", 35)
df["Age"] = pd.to_numeric(df["Age"])
df["Sales_Amount"] = pd.to_numeric(df["Sales_Amount"])
df.dtypes
```

```python
#Save the file
```

```python
df.to_csv("Cleaned_sales_dataset.csv", index=False)
```
